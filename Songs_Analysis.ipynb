# SPOTIFY & YOUTUBE DATA ANALYSIS

```python
#visualization
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("Spotify_Youtube.csv")
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.columns
```

```python
df.describe()
```

```python
extra_col=['Unnamed: 0','Uri','Url_spotify','Url_youtube']
df.drop(extra_col,axis=1,inplace=True)
```

```python
df.columns
```

```python
df.head(5)
```

#viewing the number of unique value of the each column

```python
df.nunique()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df['Views']=df['Views'].astype('int')
df['Likes']=df['Likes'].astype('int')
df['Comments']=df['Comments'].astype('int')
```

```python
df['Views'].head(10)
```

# Determining the top 10 artists on the Spotify platform whose songs have more views.


```python
df1=df.groupby('Artist')['Views'].agg('sum').reset_index().sort_values('Views',ascending=False).head(10)
```

```python
df1.columns
```

```python
df1
```

```python
df1.set_index('Artist',inplace=True)
df1.plot(kind='bar',color="g")
plt.xlabel('Artist')
plt.ylabel('Song Views')
plt.xticks(rotation=70)
plt.show()
```

# Determining the Top 10 artist with most Likes on youtube


```python
df2=df.groupby('Artist')['Likes'].agg('sum').reset_index().sort_values(by="Likes",ascending=False).head(10)
```

```python
df2
```

```python
sns.barplot(data=df2,x='Likes',y='Artist')
plt.title('Top 10 most liked Artist',fontsize="20",color = "Blue",pad='15.0' )
plt.show()
```

```python
#Determine the top 10 Streamers on spotify
```

```python
df3 = df.groupby('Artist')['Stream'].agg('sum').reset_index().sort_values('Stream',ascending=False).head(10)
```

```python
df3
```

```python
sns.barplot(data=df3,x ='Artist',y='Stream')
plt.xticks(rotation=80)
plt.title("Top 10 Streamer",pad ='20.0',fontsize=30,color='indigo')
plt.show()
```

# Determine the Relation between Acoustiness and energy

```python
plt.figure(figsize=(10,5))
sns.jointplot(data=df ,x='Acousticness',y='Energy',kind='hist')
plt.xlabel('Acoustiness',labelpad=20)
plt.ylabel('Energy',labelpad=20)
```

# Determine the Relationship between tempo and speechiness

```python
fig=plt.figure(figsize=(10,5))
sns.jointplot(data=df,x='Tempo',y='Speechiness',kind="scatter")
plt.xlabel('Speechiness',labelpad=20)
plt.ylabel('Tempo',labelpad=20)
plt.axis('tight')
```

# Relatonship between tempo and key

```python
plt.figure(figsize=(10,5))
sns.jointplot(data=df,x='Tempo',y='Key',kind='hist')
plt.xlabel('Tempo')
plt.ylabel('Key')
plt.show()
```

# Relationship bw Instrumentalness and speechiness

```python
plt.figure(figsize=(10,5))
sns.jointplot(data=df,x='Speechiness',y='Instrumentalness',kind='scatter')
plt.xlabel('Speechiness',labelpad=20)
plt.ylabel('Instrumentalness',labelpad=20)
plt.axis('tight')
```

# Relation between Instrumentalness and Acousticness

```python
sns.jointplot(data=df,x='Instrumentalness',y='Acousticness',kind='resid')
plt.show()
```
